## Process YKO library

```python
table_path='/Users/paucac/Documents/PhD/CAB/CRISPA/YKO_library/'
col_names=['Plate', 'Row','Column','ORF','Gene','Description','GO Biological Process','GO Molecular Function','GO Cellular Component','96-position']
Data = pd.read_csv(table_path+ '384format.csv', delimiter=';', names=col_names, skiprows=2)
Data = Data.drop('96-position', 1)
Data.head()
```

```python
8*2*12*2
```

```python
letter={'A':1,
        'B':2,
        'C':3,
        'D':4,
        'E':5,
        'F':6,
        'G':7,
        'H':8,
        'I':9,
        'J':10,
        'K':11,
        'L':12,
        'M':13,
        'N':14,
        'O':15,
        'P':16
        
       }
number={}

for L,N in letter.items():
    number[N]=L
```

```python
plates384={}
for i in range(1,14):
    plates384[i]=np.empty([16,24])
    
for i in range(len(Data)):
    P=Data['Plate'].iloc[i] 
    R=letter[Data['Row'].iloc[i]] -1
    C = Data['Column'].iloc[i] -1
    plates384[P][R,C] = int(i)
```

```python
#when replicating the YKO library96well plates were arranged in the following configuration:
#    Red:plateA    Blue:plateB
#.   Yellow:PlateC Green:plateD

plates96={}
for i in range(1,13+1):
        plates96['A'+str(i)]=plates384[i][::2,::2]
        plates96['B'+str(i)]=plates384[i][::2,1::2]
        plates96['C'+str(i)]=plates384[i][1::2,::2]
        plates96['D'+str(i)]=plates384[i][1::2,1::2]
        
```

```python
Data.iloc[i].values[3:]
```

```python
Data96=[]
i=0
for P in range(1,13+1):
    for L in ['A','B','C','D']:
        for row in range(1,9):
            for col in range(1,13):
                new_line= [str(P)+L,number[row], col]
                [new_line.append(cell) for cell in list( Data.iloc[i].values[3:])]
                #print(new_line)
                Data96.append(new_line)
                i+=1
        
Data96=np.vstack(Data96)
```

```python
gene_plate =plates96['A1'].copy()
gene_plate = (gene_plate*0).astype(str)
```

```python
col_names=['Strain ID','Plate.','Row.','Col.','ORF.','Gene.','ts allele','ts on SC','Background','Description','Genotype Mata','Plate','Row','Column','ORF','Gene','Decription','TS allele']

TS_384 = pd.read_csv(table_path+ 'TS_library.csv', names=col_names, delimiter=';', skiprows=1)
TS_384 = TS_384.drop(['Plate.','Row.','Col.','ORF.','Gene.'], 1)
TS_384.head(4)
```

```python
plates384TS={}
for i in range(1,4):
    plates384TS[i]=np.empty([16,24])
    
for i in range(len(TS_384)):
    P=TS_384['Plate'].iloc[i] 
    R=letter[TS_384['Row'].iloc[i]] -1
    C = TS_384['Column'].iloc[i] -1
    plates384TS[P][R,C] = int(i)
```

```python
plate=1
plates384TS[plate][3][0], plates96TS['D'+str(plate)][1][0]
```

```python
#when replicating the TS library96well plates were arranged in the following configuration:
#    Red:plateA    Blue:plateB
#.   Yellow:PlateD Green:plateC
plates96TS={}
for i in range(1,3+1):
        plates96TS['A'+str(i)]=plates384TS[i][::2,::2]
        plates96TS['B'+str(i)]=plates384TS[i][::2,1::2]
        plates96TS['D'+str(i)]=plates384TS[i][1::2,::2]
        plates96TS['C'+str(i)]=plates384TS[i][1::2,1::2]
```

```python
Data96TS=[]
i=0
for P in range(1,3+1):
    for L in ['A','B','C','D']:
        for row in range(1,9):
            for col in range(1,13):
                i=int(plates96TS[L+str(P)][row-1][col-1])
                new_line= []
                [new_line.append(cell) for cell in list(TS_384.iloc[i].values[0:6])]
                [new_line.append(cell) for cell in [str(P)+L,number[row], col]]
                [new_line.append(cell) for cell in list(TS_384.iloc[i].values[9:])]
                #print(new_line)
                Data96TS.append(new_line)

        
Data96TS=np.vstack(Data96TS)

Data96TS=pd.DataFrame(Data96TS, columns = list(TS_384.columns))

Data96TS.head(-3)
```
